## Testing Titanic Dataset with Tensorflow Data Validator

```python
import tensorflow_data_validation as tfdv
```

```python
TRAIN_DATA = '/home/saradindu/dev/GDG_Cloud_Community_Day_Aug07_2022/notebooks/train.csv'
TEST_DATA = '/home/saradindu/dev/GDG_Cloud_Community_Day_Aug07_2022/notebooks/test.csv'
```

```python
train_stats = tfdv.generate_statistics_from_csv(TRAIN_DATA, delimiter=',')
```

```python
tfdv.visualize_statistics(train_stats)
```

```python
schema = tfdv.infer_schema(train_stats)
```

```python
tfdv.display_schema(schema)
```

## Validation

```python
new_csv_stats = tfdv.generate_statistics_from_csv(TEST_DATA, delimiter=',')
```

```python
anomalies = tfdv.validate_statistics(statistics=new_csv_stats, schema=schema)
```

```python
tfdv.display_anomalies(anomalies)
```
